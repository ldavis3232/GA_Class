###Preliminary Questions
I wanted to understand if the demographics of the union membership influence the likelihood they will attend union events. For example, what is the relationship between member age and total union activities attended?  I had three preliminary questions. 

1.	What are the basic summary statistics of my dataset?
2.	How are the features of my dataset distributed? 
3.	Is there a strong correlation between my target variables (**Total_Activities,  5Activites, and 1Activity**) with other features of the dataset?


###The Dataset
The **UnionData** dataset represents the membership demographics for a property service labor union.  The dataset was transformed per **Tidy Data** standards.  This means the data was transformed so that each column/ feature represents a variable **(demographic)** and row represents an observation **(member)**.  

###Summary
After the dataset was cleaned and transformed the dataset was described through summary statistics.  Several loops were used to establish frequency by feature. The dataset was then spilt into various groups.  Then aggregate summary statistics were applied to each group (sums, means, and counts).   Next the distribution of each demographic was visualized.  Then the target variable to feature relationship was explored by visualizing the distribution of the target variable based on demographic s. Then feature to feature relationships were explored through scatter visualization. Finally, the results were summarized by findings.
