```python
#Running a ols regression with windspeed and temp against causal
import statsmodels.formula.api as smf
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(data.temp, data.casual, 'g.')

model = smf.ols(formula='casual ~ temp', data=data)
results = model.fit()
print 'NORMAL FIT SUMMARY'
print(results.summary())
print

axes[1].plot(data.windspeed, data.casual, 'm.')

log_model = smf.ols(formula='casual ~ windspeed', data=data)
log_results = log_model.fit()
print 'LOG-LOG FIT SUMMARY'
print(log_results.summary())

print fig
```

###Results Summary
The R value (correlation coefficient) of temp and wind speed ols regressions are **0.211 and .008** respectively. The R value of temp suggests a weak uphill relationship and wind speed suggest almost no linear relationship. However, this could be due to the large band/ scatter of both. The P values of **0.0** suggest that these target variables are meaningful addition to the model. The slope coefficients (**117 and 36 **) suggest statistical significant for the relationship with causal riders. Especially, in temp.

####Business Application
Despite the weak R values I feel confidently because of the visualizations, slope coefficients, and P values on the relationships with temperature and wind speed on casual riders. As the temperature increases the amount of casual riders dramatically increase. Whereas when the wind speed increases the amount of casual riders decrease. I would make the simple business suggestion that on warm days Capital Bikshare increases its inventory of available bikes to casual riders and on windy days do the inverse.
